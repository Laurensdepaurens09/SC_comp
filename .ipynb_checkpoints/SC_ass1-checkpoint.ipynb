```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python

def wave_eq(t,L = 1.,N = 100, dx = .01,dt = .001): 
    """ Function integrating time for 2 demensianal wave function. 
    """
    
    #Some constant
    c = 1
    
    # intilize string 
    string=np.zeros([N,t])
    # time 0 : 
    for x in range(N):
        string[x,0]=math.sin(2*math.pi*(x/float(N-1)))
        string[x,1]=math.sin(2*math.pi*(x/float(N-1)))

    # loop over all time steps : 
    for t in range(1,999):
        for x in range(1,99):
            string[x,t+1]=(((c*dt)/dx)**2)*(string[x-1,t]+string[x+1,t]-2*string[x,t])-string[x,t-1]+2*string[x,t]
    
    # return subset of output:
    for i in range(t):
        if i%10==0:
            plt.plot(string[:,i])
    plt.show()
    return string
```

```python
st = wave_eq(1000)
```
